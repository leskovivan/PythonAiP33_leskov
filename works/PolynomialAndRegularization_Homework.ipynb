# Поліноміальна регресія та регуляризація
## Завдання 1
Побудуйте модель поліноміальної регресії для однієї ознаки на базі датасету California Housing.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score

# Завантаження датасету
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
y = california.target

# Використовуємо лише MedInc
X1 = df[['MedInc']]

X_train1, X_test1, y_train, y_test = train_test_split(X1, y, test_size=0.2, random_state=42)

poly2 = PolynomialFeatures(degree=2)
X_train_poly2 = poly2.fit_transform(X_train1)
X_test_poly2 = poly2.transform(X_test1)

model_poly2 = LinearRegression()
model_poly2.fit(X_train_poly2, y_train)
r2_test_poly2 = r2_score(y_test, model_poly2.predict(X_test_poly2))

print(f"R² на тестовій вибірці (ступінь 2): {r2_test_poly2:.5f}")
```

## Завдання 2
Перевірте, як ступінь полінома впливає на якість моделі.

```python
degrees = [1, 2, 3]
results2 = []

for d in degrees:
    poly = PolynomialFeatures(degree=d)
    X_tr_p = poly.fit_transform(X_train1)
    X_te_p = poly.transform(X_test1)
    
    model = LinearRegression()
    model.fit(X_tr_p, y_train)
    
    r2_train = r2_score(y_train, model.predict(X_tr_p))
    r2_test = r2_score(y_test, model.predict(X_te_p))
    
    results2.append({
        'Ступінь полінома': d,
        'R² поїзд': r2_train,
        'Тест R²': r2_test
    })

df_task2 = pd.DataFrame(results2)
display(df_task2)
```

## Завдання 3
Застосуйте Lasso-регуляризацію і порівняйте зі звичайною лінійною регресією.

```python
features = ['MedInc', 'HouseAge', 'AveRooms', 'AveBedrms']
X3 = df[features]

X_train3, X_test3, y_train, y_test = train_test_split(X3, y, test_size=0.2, random_state=42)

# Звичайна лінійна регресія
lr = LinearRegression()
lr.fit(X_train3, y_train)
lr_r2 = r2_score(y_test, lr.predict(X_test3))

# Lasso регуляризація
lasso = Lasso(alpha=0.1)
lasso.fit(X_train3, y_train)
lasso_r2 = r2_score(y_test, lasso.predict(X_test3))

zero_coefs_lasso = np.sum(lasso.coef_ == 0)
total_coefs = len(lasso.coef_)

print(f"Linear Regression R²: {lr_r2:.5f}")
print(f"Lasso (alpha=0.1) R²: {lasso_r2:.5f}")
print(f"Кількість коефіцієнтів, що стали 0 у Lasso: {zero_coefs_lasso} з {total_coefs}")
```

## Завдання 4
Застосуйте Ridge-регуляризацію і порівняйте зі звичайною лінійною регресією.

```python
ridge = Ridge(alpha=1.0)
ridge.fit(X_train3, y_train)
ridge_r2 = r2_score(y_test, ridge.predict(X_test3))

zero_coefs_lr = np.sum(lr.coef_ == 0)
zero_coefs_ridge = np.sum(ridge.coef_ == 0)

results4 = pd.DataFrame({
    'Модель': ['LinearRegression', 'Ridge(alpha=1.0)'],
    'Тест R²': [lr_r2, ridge_r2],
    'Кількість ненульових коеф.': [
        len(lr.coef_) - zero_coefs_lr,
        len(ridge.coef_) - zero_coefs_ridge
    ]
})
display(results4)
```
