```python
import matplotlib.pyplot as plt
import numpy as np

ages = np.random.randint(18, 71, size=100)

mean_age = np.mean(ages)

plt.hist(ages, bins=10, color='skyblue', edgecolor='black')
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label=f'Середній вік: {mean_age:.1f}')

plt.title('Розподіл віку 100 осіб')
plt.xlabel('Вік')
plt.ylabel('Кількість осіб')
plt.legend()

plt.show()
```

## Завдання 3
Є результати іспитів за трьома групами студентів. Потрібно порівняти розкид оцінок між групами.

```python
import matplotlib.pyplot as plt
import numpy as np

group_A = np.random.randint(60, 100, size=20)
group_B = np.random.randint(50, 95, size=20)
group_C = np.random.randint(70, 100, size=20)

data = [group_A, group_B, group_C]
labels = ['Група A', 'Група B', 'Група C']

plt.boxplot(data, labels=labels, patch_artist=True)

plt.title('Розкид оцінок між групами студентів')
plt.xlabel('Групи')
plt.ylabel('Оцінки (в балах)')
plt.grid(axis='y', linestyle='--')

plt.show()
```

## Завдання 4
Є денні значення температури та вологості за тиждень. Потрібно показати обидві метрики на загальній діаграмі.
Створіть список дат (7 днів) і два списки значень: температура і вологість.
Побудуйте лінійний графік температури.
На тій самій області побудуйте графік вологості.
Додайте легенду, підписи осей і заголовок.
Поверніть підписи дат на осі X для читабельності та відобразіть графік.

```python
import matplotlib.pyplot as plt

days = ['Понеділок', 'Вівторок', 'Середа', 'Четвер', 'П\'ятниця', 'Субота', 'Неділя']
temperatures = [22, 24, 25, 23, 26, 28, 27]
humidities = [50, 48, 45, 55, 42, 40, 43]

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('Дні тижня')
ax1.set_ylabel('Температура (°C)', color=color)
ax1.plot(days, temperatures, color=color, label='Температура', marker='o')
ax1.tick_params(axis='y', labelcolor=color)

plt.xticks(rotation=45)

ax2 = ax1.twinx()  
color = 'tab:blue'
ax2.set_ylabel('Вологість (%)', color=color)  
ax2.plot(days, humidities, color=color, label='Вологість', marker='s')
ax2.tick_params(axis='y', labelcolor=color)

plt.title('Денні значення температури та вологості за тиждень')
fig.tight_layout() 
plt.show()
```

## Завдання 6
Дано чотири набори метрик продукту: конверсія, утримання, середній чек, кількість замовлень. Потрібно показати їх на окремих панелях.
Створіть чотири набори числових значень за однією шкалою часу.
Створіть сітку з чотирьох підграфіків 2×2.
На кожному підграфіку побудуйте відповідний графік.
Додайте заголовки для кожного підграфіка та загальний заголовок.
Відобразіть результат на екрані.

```python
import matplotlib.pyplot as plt
import numpy as np

months = np.arange(1, 13) 

conversion = np.random.uniform(2.0, 5.0, size=12) 
retention = np.random.uniform(30.0, 60.0, size=12)  
avg_check = np.random.randint(500, 1500, size=12)  
orders = np.random.randint(1000, 5000, size=12) 

fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('Аналітика продукту за рік', fontsize=16)

axs[0, 0].plot(months, conversion, color='blue', marker='o')
axs[0, 0].set_title('Конверсія (%)')
axs[0, 0].set_xticks(months)
axs[0, 0].grid(True)

axs[0, 1].plot(months, retention, color='green', marker='s')
axs[0, 1].set_title('Утримання (%)')
axs[0, 1].set_xticks(months)
axs[0, 1].grid(True)

axs[1, 0].plot(months, avg_check, color='orange', marker='^')
axs[1, 0].set_title('Середній чек (грн)')
axs[1, 0].set_xticks(months)
axs[1, 0].grid(True)

axs[1, 1].plot(months, orders, color='red', marker='D')
axs[1, 1].set_title('Кількість замовлень')
axs[1, 1].set_xticks(months)
axs[1, 1].grid(True)

for ax in axs.flat:
    ax.set(xlabel='Місяць', ylabel='Значення')

for ax in axs.flat:
    ax.label_outer()

plt.tight_layout() 
plt.show()
```
