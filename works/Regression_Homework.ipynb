# Базова регресія
## Завдання 1
Побудуйте модель лінійної регресії на одній ознаці та порахуйте якість.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error

# Завантаження датасету
california = fetch_california_housing()
df = pd.DataFrame(california.data, columns=california.feature_names)
df['Target'] = california.target

# Вибірка 5000 випадкових рядків
df_subset = df.sample(n=5000, random_state=42)

X1 = df_subset[['MedInc']]
y1 = df_subset['Target']

X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y1, test_size=0.2, random_state=42)

model1 = LinearRegression()
model1.fit(X_train1, y_train1)
r2_1 = r2_score(y_test1, model1.predict(X_test1))

res1 = pd.DataFrame({
    'Модель': ['LinearRegression'],
    'Ознака': ['MedInc'],
    'R²': [r2_1]
})
display(res1)
```

## Завдання 2
Порівняйте якість моделі на одній ознаці та на двох ознаках.

```python
X2 = df_subset[['MedInc', 'AveRooms']]
y2 = df_subset['Target']

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size=0.2, random_state=42)

model2 = LinearRegression()
model2.fit(X_train2, y_train2)
r2_2 = r2_score(y_test2, model2.predict(X_test2))

res2 = pd.DataFrame({
    'Модель': ['LinearRegression', 'LinearRegression'],
    'Ознаки': ['MedInc', 'MedInc, AveRooms'],
    'R²': [r2_1, r2_2]
})
display(res2)
```

## Завдання 3
Обчисліть кілька метрик помилки для однієї моделі (на двох ознаках).

```python
y_pred2 = model2.predict(X_test2)

mae = mean_absolute_error(y_test2, y_pred2)
mse = mean_squared_error(y_test2, y_pred2)
rmse = np.sqrt(mse)
r2_3 = r2_score(y_test2, y_pred2)

res3 = pd.DataFrame({
    'Метрика': ['MAE', 'MSE', 'RMSE', 'R²'],
    'Значення': [mae, mse, rmse, r2_3]
})
display(res3)
```

## Завдання 4
Побудуйте графік для порівняння реальних і передбачених значень.

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test2, y_pred2, alpha=0.5)
plt.plot([y_test2.min(), y_test2.max()], [y_test2.min(), y_test2.max()], 'r--', lw=2)
plt.xlabel('Реальні значення (y_test)')
plt.ylabel('Передбачені значення (y_pred)')
plt.title('Порівняння реальних і передбачених значень')
plt.savefig('scatter_plot.png')
plt.show()
```
