## Завдання 1: Робота з даними про книги

```python
import pandas as pd

data_books = {
    'Назва': ['Кобзар', '1984', 'Майстер і Маргарита', 'Колгосп тварин', 'Маленький принц', 'Алхімік'],
    'Автор': ['Тарас Шевченко', 'Джордж Орвелл', 'Михайло Булгаков', 'Джордж Орвелл', 'Антуан де Сент-Екзюпері', 'Пауло Коельо'],
    'Рік видання': [1840, 1949, 1967, 1945, 1943, 1988],
    'Ціна': [250, 300, 350, 280, 200, 220]
}
books_df = pd.DataFrame(data_books)

print("Вихідний DataFrame:")
print(books_df)

avg_price = books_df['Ціна'].mean()
print(f"\nСередня ціна книг: {avg_price:.2f} грн")

books_after_2015 = books_df[books_df['Рік видання'] > 2015]
print("\nКниги, видані після 2015 року:")
print(books_after_2015) # Примітка: у вихідних даних таких книг немає, тому результат буде порожнім. Давайте додамо одну.
new_book = pd.DataFrame([{'Назва': 'Нова книга', 'Автор': 'Новий Автор', 'Рік видання': 2020, 'Ціна': 400}])
books_df = pd.concat([books_df, new_book], ignore_index=True)
books_after_2015 = books_df[books_df['Рік видання'] > 2015]
print("\nКниги, видані після 2015 року (після додавання нової книги):")
print(books_after_2015)


sorted_books_df = books_df.sort_values(by='Ціна', ascending=True)
print("\nDataFrame, відсортований за ціною:")
print(sorted_books_df)
```

## Завдання 2: Аналіз замовлень з CSV-файлу

```python
import pandas as pd
import os

# Створення зразкового CSV-файлу
orders_data = {
    'Номер замовлення': range(1, 11),
    'Клієнт': ['Іван', 'Марія', 'Петро', 'Іван', 'Олена', 'Марія', 'Іван', 'Анна', 'Петро', 'Олена'],
    'Дата': pd.to_datetime(['2023-01-15', '2023-01-16', '2023-01-17', '2023-01-18', '2023-01-19', '2023-01-20', '2023-01-21', '2023-01-22', '2023-01-23', '2023-01-24']),
    'Сума': [1500, 2500, 800, 1200, 3000, 1800, 2200, 900, 1100, 1400]
}
orders_df = pd.DataFrame(orders_data)
orders_df.to_csv('orders.csv', index=False)

# Читання даних з файлу
read_orders_df = pd.read_csv('orders.csv')

print("Перші 10 рядків DataFrame:")
print(read_orders_df.head(10))

client_orders_count = read_orders_df['Клієнт'].value_counts()
print("\nКількість замовлень кожного клієнта:")
print(client_orders_count)

max_order = read_orders_df['Сума'].max()
min_order = read_orders_df['Сума'].min()
print(f"\nМаксимальна сума замовлення: {max_order}")
print(f"Мінімальна сума замовлення: {min_order}")

total_sum = read_orders_df['Сума'].sum()
print(f"\nЗагальна сума всіх замовлень: {total_sum}")

# Видалення створеного файлу
os.remove('orders.csv')
```

## Завдання 3: Аналіз даних про продукти харчування

```python
import pandas as pd

food_data = {
    'Продукт': ['Яблуко', 'Банан', 'Курка', 'Рис', 'Шоколад', 'Горіхи', 'Йогурт', 'Хліб', 'Сир', 'Риба'],
    'Категорія': ['Фрукти', 'Фрукти', 'М\'ясо', 'Крупи', 'Солодощі', 'Горіхи', 'Молочні', 'Випічка', 'Молочні', 'Риба'],
    'Калорії': [52, 89, 239, 130, 546, 654, 59, 265, 402, 206],
    'Білки': [0.3, 1.1, 27, 2.7, 4.9, 20, 10, 9, 25, 22]
}
food_df = pd.DataFrame(food_data)

print("Вихідний DataFrame:")
print(food_df)

high_calorie_food = food_df[food_df['Калорії'] > 300]
print("\nПродукти з калорійністю вище 300:")
print(high_calorie_food)

avg_protein_by_category = food_df.groupby('Категорія')['Білки'].mean()
print("\nСередня кількість білків за категоріями:")
print(avg_protein_by_category)

sorted_food_df = food_df.sort_values(by='Калорії', ascending=False)
print("\nDataFrame, відсортований за калорійністю:")
print(sorted_food_df)
```

## Завдання 4: Аналіз даних про співробітників та проекти

```python
import pandas as pd

employees_data = {
    'Ім\'я': ['Анна', 'Богдан', 'Вікторія', 'Григорій', 'Анна', 'Богдан', 'Вікторія', 'Григорій'],
    'Проект': ['Альфа', 'Бета', 'Альфа', 'Гамма', 'Бета', 'Гамма', 'Альфа', 'Бета'],
    'Години': [40, 35, 45, 50, 20, 30, 25, 40]
}
employees_df = pd.DataFrame(employees_data)

print("Вихідний DataFrame:")
print(employees_df)

total_hours_employee = employees_df.groupby('Ім\'я')['Години'].sum()
print("\nЗагальна кількість годин за кожним співробітником:")
print(total_hours_employee)

total_hours_project = employees_df.groupby('Проект')['Години'].sum()
print("\nЗагальна кількість годин за кожним проектом:")
print(total_hours_project)

hardest_working_employee = total_hours_employee.idxmax()
print(f"\nСпівробітник, який витратив найбільше годин: {hardest_working_employee}")
```

## Завдання 5: Аналіз продажів квитків

```python
import pandas as pd

tickets_data = {
    'Фільм': ['Дюна', 'Інтерстеллар', 'Барбі', 'Оппенгеймер', 'Дюна', 'Барбі', 'Інтерстеллар', 'Оппенгеймер', 'Дюна', 'Барбі', 'Інтерстеллар', 'Оппенгеймер'],
    'Місто': ['Київ', 'Львів', 'Київ', 'Харків', 'Львів', 'Харків', 'Київ', 'Львів', 'Харків', 'Київ', 'Львів', 'Харків'],
    'Продані квитки': [150, 120, 200, 180, 130, 190, 160, 170, 140, 210, 110, 190]
}
tickets_df = pd.DataFrame(tickets_data)

print("Вихідний DataFrame:")
print(tickets_df)

total_tickets_film = tickets_df.groupby('Фільм')['Продані квитки'].sum()
print("\nЗагальна кількість проданих квитків за кожним фільмом:")
print(total_tickets_film)

total_tickets_city = tickets_df.groupby('Місто')['Продані квитки'].sum()
print("\nЗагальна кількість квитків по кожному місту:")
print(total_tickets_city)

top_selling_film = total_tickets_film.idxmax()
print(f"\nФільм із найбільшою кількістю продажів: {top_selling_film}")
```

## Завдання 6: Об'єднання даних про клієнтів та замовлення

```python
import pandas as pd

clients_data = {
    'ID': [1, 2, 3, 4, 5],
    'Ім\'я': ['Іван', 'Марія', 'Петро', 'Олена', 'Анна'],
    'Місто': ['Київ', 'Львів', 'Харків', 'Київ', 'Одеса']
}
clients_df = pd.DataFrame(clients_data)

orders_data = {
    'ID клієнта': [3, 1, 2, 4, 1, 5, 3],
    'Замовлення': ['Книга', 'Телефон', 'Навушники', 'Чохол', 'Зарядка', 'Мишка', 'Клавіатура'],
    'Вартість': [350, 25000, 1500, 500, 700, 800, 2000]
}
orders_df = pd.DataFrame(orders_data)

print("DataFrame з клієнтами:")
print(clients_df)
print("\nDataFrame з замовленнями:")
print(orders_df)

merged_df = pd.merge(clients_df, orders_df, left_on='ID', right_on='ID клієнта')
print("\nОб'єднаний DataFrame:")
print(merged_df)

total_cost_per_client = merged_df.groupby(['ID', 'Ім\'я', 'Місто'])['Вартість'].sum().reset_index()
print("\nЗагальна вартість замовлень за кожним клієнтом:")
print(total_cost_per_client)
```
