## Завдання 1


```python
import numpy as np

arr = np.random.randint(0, 51, 20)
print("Вихідний масив:", arr)

threshold = int(input("Введіть порогове значення: "))

count_greater = np.sum(arr > threshold)
print(f"Кількість елементів більших за {threshold}: {count_greater}")

max_val = np.max(arr)
max_idx = np.argmax(arr)
print(f"Максимальне значення: {max_val}, позиція першої появи (індекс): {max_idx}")

sorted_arr = np.sort(arr)[::-1]
print("Відсортований масив (за спаданням):", sorted_arr)
```

## Завдання 2


```python
import numpy as np

low = int(input("Введіть нижню межу діапазону: "))
high = int(input("Введіть верхню межу діапазону: "))

matrix = np.random.randint(low, high + 1, (5, 5))
print("\nВихідна матриця 5x5:\n", matrix)

main_diag = np.diag(matrix)
print("\nГоловна діагональ:", main_diag)

diag_sum = np.sum(main_diag)
print("Сума елементів головної діагоналі:", diag_sum)

matrix_lower = np.tril(matrix)
print("\nМатриця після обнулення елементів вище головної діагоналі:\n", matrix_lower)
```

## Завдання 3

```python
import numpy as np

start = int(input("Введіть початок діапазону: "))
end = int(input("Введіть кінець діапазону: "))

matrix_6x5 = np.random.randint(start, end + 1, (6, 5))
print("\nМатриця 6x5:\n", matrix_6x5)

row_sums = np.sum(matrix_6x5, axis=1)
print("\nСума елементів у кожному рядку:", row_sums)

col_maxs = np.max(matrix_6x5, axis=0)
print("Максимальні значення за кожним стовпчиком:", col_maxs)
```

## Завдання 4


```python
import numpy as np

low = int(input("Введіть нижню межу (напр., -20): "))
high = int(input("Введіть верхню межу (напр., 20): "))

arr_15 = np.random.randint(low, high + 1, 15)
print("\nВихідний масив:", arr_15)

neg_arr = arr_15[arr_15 < 0]
print("Від'ємні елементи:", neg_arr)

mod_arr = np.where(arr_15 < 0, 0, arr_15)
print("Змінена копія (від'ємні замінено на 0):", mod_arr)

zero_count = np.sum(mod_arr == 0)
print("Кількість нульових елементів у зміненому масиві:", zero_count)
```

## Завдання 5

```python
import numpy as np

length = int(input("Введіть довжину масивів: "))

arr1 = np.random.randint(0, 11, length)
arr2 = np.random.randint(10, 21, length)

print("\nПерший масив (0-10):", arr1)
print("Другий масив (10-20):", arr2)

combined_arr = np.concatenate((arr1, arr2))
print("\nОб'єднаний масив:", combined_arr)

sum_arr = arr1 + arr2
diff_arr = arr1 - arr2
print("\nПоелементне додавання:", sum_arr)
print("Поелементна різниця (перший - другий):", diff_arr)
```

## Завдання 6


```python
import numpy as np

rows = int(input("Введіть кількість рядків вихідної матриці: "))
cols = int(input("Введіть кількість стовпців вихідної матриці: "))

total_elements = rows * cols
matrix_seq = np.arange(total_elements).reshape(rows, cols)
print("\nВихідна матриця:\n", matrix_seq)

new_rows = int(input("\nВведіть нову кількість рядків: "))
new_cols = int(input("Введіть нову кількість стовпців: "))

if total_elements != (new_rows * new_cols):
    print(f"Помилка: неможливо перетворити матрицю з {total_elements} елементів у форму {new_rows}x{new_cols}.")
else:
    new_matrix = matrix_seq.reshape(new_rows, new_cols)
    print("\nПеретворена матриця:\n", new_matrix)

    row_mins = new_matrix.min(axis=1)
    row_maxs = new_matrix.max(axis=1)
    print("\nМінімальні значення в кожному рядку:", row_mins)
    print("Максимальні значення в кожному рядку:", row_maxs)

    total_sum = new_matrix.sum()
    print("\nЗагальна сума всіх елементів нової матриці:", total_sum)
```
