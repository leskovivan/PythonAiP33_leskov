```python
from IPython.display import display, Markdown

temps = [2, 5, 4, 6, 7, 3, 1]


t_mean = sum(temps) / len(temps)
t_max = max(temps)
t_min = min(temps)
t_diff = t_max - t_min

markdown_table = f"""
### Результати аналізу температур
| Показник | Значення |
| :--- | :--- |
| Середнє значення | {t_mean:.2f} |
| Максимальне значення | {t_max} |
| Мінімальне значення | {t_min} |
| Різниця (Макс - Мін) | {t_diff} |
"""


display(Markdown(markdown_table))
```

```python
from IPython.display import display, Markdown


expenses = [1200, 1500, 800, 600, 1000, 400, 900]

total = sum(expenses)
mean_exp = total / len(expenses)
max_exp = max(expenses)
min_exp = min(expenses)

days_over_1000 = sum(1 for exp in expenses if exp > 1000)

markdown_table = f"""
### Аналіз щоденних витрат
| Показник | Значення |
| :--- | :--- |
| Загальна сума | {total} |
| Середнє значення | {mean_exp:.2f} |
| Максимальна витрата | {max_exp} |
| Мінімальна витрата | {min_exp} |
| Кількість днів із витратами > 1000 | {days_over_1000} |
"""

display(Markdown(markdown_table))
```

```python
from collections import Counter
from IPython.display import display, Markdown


votes = ["apple", "banana", "apple", "orange", "banana", "apple", "apple", "apple", "orange"]


counts = Counter(votes)

leader = counts.most_common(1)[0][0]
top_3 = counts.most_common(3)


markdown_table = "### Результати голосування\n"
markdown_table += "| Фрукт | Кількість голосів |\n| :--- | :--- |\n"

for fruit, count in counts.items():
    markdown_table += f"| {fruit} | {count} |\n"

markdown_table += f"\n**Фрукт-лідер:** {leader}\n"

markdown_table += "\n**Топ-3:**\n"
for idx, (fruit, count) in enumerate(top_3, 1):
    markdown_table += f"{idx}. {fruit} ({count} голосів)\n"

display(Markdown(markdown_table))
```

```python
from IPython.display import display, Markdown

# 1. Списки оцінок
math = [80, 92, 75, 60, 88]
physics = [70, 65, 90, 85, 78]
english = [95, 100, 85, 90, 87]

subjects = {
    "Математика": math,
    "Фізика": physics,
    "Англійська": english
}

results = {}
for subject, scores in subjects.items():
    results[subject] = {
        "mean": sum(scores) / len(scores),
        "min": min(scores),
        "max": max(scores)
    }


best_subject = max(results, key=lambda sub: results[sub]["mean"])


markdown_table = "### Аналіз оцінок за предметами\n"
markdown_table += "| Предмет | Середній бал | Мінімальний бал | Максимальний бал |\n| :--- | :--- | :--- | :--- |\n"

for subject, stats in results.items():
    markdown_table += f"| {subject} | {stats['mean']:.1f} | {stats['min']} | {stats['max']} |\n"

markdown_table += f"\n**Предмет з найвищим середнім результатом:** {best_subject}"

display(Markdown(markdown_table))
```
