### Звіт із продажів
У цьому розділі ми рахуємо підсумки за простими даними

```python
sales = [120, 135, 100, 150, 90, 200, 170]
print(sales)
```

```python
from IPython.display import Markdown
```

#### Зразок таблиці

|Title 1|Title 2|
|-|-|
|value 1-1|value 2-1|
|value 1-2|value 2-2|

#### Зразок HTML таблиці
<table>
<thead>
<tr><th>Header</th><th>Header</th><th>Header</th></tr>
</thead>
<tbody>
<tr><td>A</td><td>B</td><td>C</td></tr>
<tr><td>X</td><td>Y</td><td>Z</td></tr>
</tbody>
</table>

```python
Markdown("Показник|Значення|\n|-|-|\n|Загальна сума|%.2f|\n|Середнє|%.2f|\n|Максимум|%.2f|\n|Мінімум|%.2f|" % (
    sum(sales),
    sum(sales) / len(sales),
    max(sales),
    min(sales)
))
```
