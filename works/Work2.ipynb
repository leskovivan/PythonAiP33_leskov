```python
print("Форма масиву: ", rand_3_5.shape)
```

```python
trans = np.transpose(rand_3_5)
print("Транспонування:\n", trans)
print("Форма транспонованого масиву: ", trans.shape)
```

```python
resh = np.reshape(rand_3_5, shape=(5,3))
print("Зміна розмірності (решейпінг):\n", resh)
print("Форма нового масиву: ", resh.shape)
# дія решейпінгу - серіалізація вихідного масиву 
# (перенесення в один вимір один-за-одним)
# та наступний переділ згідно з заданими
# розмірностями 
```

## Завдання 4
* Створіть масив розміром 10×10, заповнений випадковими числами з плаваючою крапкою від 0 до 1.
* Знайдіть мінімальне та максимальне значення в усьому масиві.
* Визначте індекси мінімального та максимального елементів
(наприклад: (рядок, стовпець)).
* Виведіть сам масив і окремо знайдені значення разом з їхніми індексами.

```python
rand_10_10 = np.random.rand(10,10)  # виведення великих об'ємів незручне
np.set_printoptions(precision=4)    # Тільки для виведення можна обмежити
print(*rand_10_10, sep='\n')        # точність (тільки для масивів)
```

```python
max_v = np.max(rand_10_10)
print("Максимальне значення у масиві: ", max_v)
max_i = np.argmax(rand_10_10)  # індекс суцільний - від самого початку масиву
print("Знаходиться за індексом: ", max_i)
# перераховуємо суцільний індекс на набір індексів по заданій формі масиву
max_coord = np.unravel_index(max_i, rand_10_10.shape)
print("Що відповідає сукупності індексів:", ", ".join(map(str, max_coord)))
```

```python
min_v = np.min(rand_10_10)
print("Максимальне значення у масиві: ", min_v)
min_i = np.argmin(rand_10_10) 
print("Знаходиться за індексом: ", min_i)
min_coord = np.unravel_index(min_i, rand_10_10.shape)
print("Що відповідає сукупності індексів:", ", ".join(map(str, min_coord)))
```

## Завдання 5
* Створіть одновимірний масив із 15 випадкових цілих чисел від -10 до 10.
* Виведіть масив.
* За допомогою булевої індексації виберіть тільки додатні елементи.
* За допомогою булевої індексації виберіть тільки парні елементи.
* Виведіть обидва результати на екран.

```python
rand_15 = np.random.randint(-10, 10, size=15)
print(rand_15)
```

```python
res_pos = rand_15 > 0
print("Булева індексація дозволяє застосувати умову до масиву і одержати масив результатів:\n",
      res_pos)
print("Використання масиву булевих значень як індексу залишає лише ті елементи, для яких True:\n",
      rand_15[res_pos])
print("Кількість додатніх елементів:", np.count_nonzero(res_pos))
```

```python
print("Парні значення:", rand_15[rand_15 % 2 == 0])
```
